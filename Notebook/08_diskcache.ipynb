### creating a cache manager


```python
from diskcache import Cache
from hashlib import sha256

# Create cache directory
cache = Cache("./cache")


def get_cache_key(prompt: str) -> str:
    """Generate a unique cache key."""
    return sha256(prompt.strip().lower().encode()).hexdigest()


def get_cached_response(prompt: str):
    key = get_cache_key(prompt)
    return cache.get(key)


def save_response(prompt: str, response: str):
    key = get_cache_key(prompt)
    cache.set(key, response, expire=86400)  # Cache for 24 hours
```

### using diskcache to cache responses from the LLM


```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="llama3.2:latest",
    temperature=0.3
)

prompt = "write a 1000 word essay on the importance of AI in modern society."

# Check cache first
cached = get_cached_response(prompt)

if cached:
    print("⚡ Response from Cache\n")
    print(cached)

else:
    print("🤖 Calling LLM...\n")

    response = llm.invoke(prompt)

    # Save response
    save_response(prompt, response.content)

    print(response.content)
```
